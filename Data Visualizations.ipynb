```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
titanic = sns.load_dataset("titanic")
iris = sns.load_dataset("iris")
tips = sns.load_dataset("tips")
transactions = pd.read_csv("transactions.csv")
```

```python
plt.figure(figsize=(14,6))
sns.lineplot(data=tips, x='day', y='total_bill')
plt.title("Line Chart: total_bill vs day")
plt.show()
```

```python
plt.figure(figsize=(14,7))
sns.scatterplot(data=transactions, x="sales_amount", y="profit_margin")
plt.show()
```

```python
# plt.figure(figsize=(10,5))
sns.displot(data=iris, x="sepal_length", color='green', kind='hist')
plt.show()
```

```python
plt.figure(figsize=(15,7))
sns.kdeplot(data=iris, x="sepal_length", color='green')
plt.show()
```

```python
sns.histplot(data=iris, x="sepal_width", color="blue")
plt.show()
```

```python
# plt.figure(figsize=(10, 7))
sns.catplot(data=tips, x="day", y="total_bill", hue='sex')
plt.show()
```

```python
sns.catplot(data=titanic, x="age", y="class", kind='boxen', hue='sex')
plt.show()
```

```python
plt.figure(figsize=(15,7))
sns.stripplot(data=tips, x="total_bill", y="day")
plt.show()
```

```python
plt.figure(figsize=(15,7))
sns.swarmplot(data=tips, x="total_bill", y="day")
plt.show()
```

```python
plt.figure(figsize=(14,5))
sns.boxplot(data=tips, x="total_bill", color='green')
plt.show()
```

```python
plt.figure(figsize=(12,4))
sns.boxenplot(data=tips, x='total_bill', color='yellow')
plt.show()
```

```python
plt.figure(figsize=(14,5))
sns.barplot(data=tips, x='day', y='total_bill')
plt.show()
```

```python
plt.figure(figsize=(14,6))
sns.countplot(data=transactions, x='market_code')
plt.show()
```

```python
plt.figure(figsize=(14,7))
sns.heatmap(transactions.corr())
plt.show()
```
