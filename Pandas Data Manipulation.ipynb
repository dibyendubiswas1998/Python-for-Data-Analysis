```python
# np.where(): Finding Null values.

fea1_ind = np.where(custom['fea1'].isnull() == True)
fea2_ind = np.where(custom['fea2'].isnull() == True)
fea3_ind = np.where(custom['fea3'].isnull() == True)
fea4_ind = np.where(custom['fea4'].isnull() == True)
```

```python
# map:

custom['op'] = custom['op'].map({0:False, 1:True})
custom.head()
```

```python
# drop_duplicates:

custom = custom.drop_duplicates()
custom.shape
```

```python
# Rename columns:

custom = custom.rename({'fea1': 'F1', 'fea2': 'F2', 'fea3':'F3', 'fea4':'F4', 'op': 'Output'}, axis=1)
custom.head()
```
